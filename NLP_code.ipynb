```python
import spacy

data="""Bomb: explosive device, incendiary, grenade, missile, shell
Explosive: dynamite, TNT, C4, nitroglycerin, gunpowder
Suicide bomber: martyrdom seeker, human bomb, kamikaze, suicide attacker, self-destructive
Jihad: holy war, struggle, crusade, armed struggle, fighting, conflict
Martyrdom: self-sacrifice, self-immolation, selflessness, devotion, death for a cause
Attack: assault, raid, offensive, strike, onslaught, blitz
Hijack: seize, commandeer, pirate, take over, skyjack
Weapon: armament, firepower, instrument of attack, munition, tool of violence
Militant: radical, activist, combatant, fighter, extremist
Extremist: radical, fanatic, zealot, dogmatist, ideologue
Radicalization: extremism, fanaticism, militancy, zealotry, dogmatism
Training camp: boot camp, indoctrination center, recruitment center, military training facility
Terrorist network: extremist group, radical organization, terrorist cell, insurgency, clandestine group
Ideology: belief system, doctrine, creed, philosophy, dogma
Propaganda: disinformation, misinformation, indoctrination, brainwashing, agitation
Cell: sleeper cell, clandestine group, underground organization, covert network
Safe house: hideout, refuge, sanctuary, shelter, covert location
Chemical weapon: nerve gas, mustard gas, poison gas, biochemical agent, toxic substance
Biological weapon: bioweapon, germ warfare agent, infectious agent, deadly organism
Cyberterrorism: electronic warfare, cyberwarfare, information warfare, computer terrorism, hacking terrorism""";

de=data.split("\n")
de2=[]
for i in de:
  for j in i.split(","):
    de2.append(j)
de3=[]
for i in de2:
  for j in i.split(":"):
    de3.append(j)
# print(de3)

import requests
from bs4 import BeautifulSoup
# https://en.wikipedia.org/wiki/List_of_Islamist_terrorist_attacks
url = "https://en.wikipedia.org/wiki/List_of_Islamist_terrorist_attacks"
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')

# extract all the data from the website
data = soup.get_text()



par=data
nlp = spacy.load("en_core_web_sm")
doc=nlp(par)

noun=[];person=[];verb=[]
for w in doc:
  # print(w,w.pos_)
  if w.pos_=='NOUN':
    noun.append(w.text)
  if w.pos_=='PROPN':
    person.append(w.text)

print('noun:',noun)
print('Person:',person)

import collections as c
d1=dict(c.Counter(person))
d2=dict(c.Counter(noun))

p_count=set(d1.values());v_count=set(d2.values())
p_count=sorted(p_count);v_count=sorted(v_count)


list1=[];list2=[]

print('\n\nby persons:')
for c in p_count:
  for i,j in d1.items():
    if j==c:
      for kw in de3:
        if i in kw or kw in i:
          if len(i)>2:
            list1.append(i)
            # print(i)
            break
print(list1)

print('By Nouns')
for c in v_count:
  for i,j in d2.items():
    if j==c:
      for kw in de3:
        if i in kw or kw in i:
          if len(i)>2:
            list2.append(i)
            break

print(list2)
```
